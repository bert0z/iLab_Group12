```python
import pandas as pd

# Load the dataset
file_path = '/content/parkinsons.data'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics of the dataset
summary = data.describe()

# Information about the dataset
info = data.info()

summary, info
```

```python
import matplotlib.pyplot as plt

# Plotting the distribution of the target variable (status)
plt.figure(figsize=(6, 4))
data['status'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribution of Health Status')
plt.xlabel('Health Status (0: Healthy, 1: Parkinson\'s)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Plotting distributions of features
data.hist(figsize=(25,30));

# Boxplot of a key feature grouped by health status
plt.figure(figsize=(8, 6))
data.boxplot(column='MDVP:Fo(Hz)', by='status')
plt.title('Boxplot of MDVP:Fo(Hz) Grouped by Health Status')
plt.suptitle('')
plt.xlabel('Health Status (0: Healthy, 1: Parkinson\'s)')
plt.ylabel('MDVP:Fo(Hz)')
plt.show()
```

1. Health Status Distribution:
The dataset is imbalanced, with more instances of individuals with Parkinson's disease (status = 1) compared to healthy individuals (status = 0).
2. Feature Distributions:
The distributions of features like MDVP:Fo(Hz), MDVP:Jitter(%), MDVP:Shimmer(dB), and HNR show varying patterns, indicating potential differences between healthy individuals and those with Parkinson's.
For instance, MDVP:Jitter(%) and MDVP:Shimmer(dB) seem to have a higher variance and potentially higher values in the Parkinson's group.
3. Boxplot Analysis:
The boxplot of MDVP:Fo(Hz) grouped by health status shows a difference in the median and interquartile range between the two groups, suggesting that this feature could be useful in distinguishing between healthy and Parkinson's patients.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Define the features and target variable
X_first = data.drop(columns=['name', 'status'])
y_first = data['status']

# Split the data into training and test sets
X_train_first, X_test_first, y_train_first, y_test_first = train_test_split(X_first, y_first, test_size=0.3, random_state=42)

# Train a Random Forest classifier
rf_first = RandomForestClassifier(n_estimators=100, random_state=42)
rf_first.fit(X_train_first, y_train_first)

# Get feature importances
importance_first = rf_first.feature_importances_

# Get feature names
feature_names_first = X_first.columns

# Combine the feature importances with the feature names
feature_importance_first = pd.DataFrame({'Feature': feature_names_first, 'Importance': importance_first})

# Sort the feature importances
feature_importance_first = feature_importance_first.sort_values(by='Importance', ascending=False)

feature_importance_first
```

### **Top Features:**

1.   PPE (Pitch Period Entropy): This is the most important feature, indicating that variations in the fundamental frequency play a significant role in distinguishing between healthy individuals and those with Parkinson's disease.
2.   spread1: This feature, which is related to the non-linear variation in the fundamental frequency, is also highly important.
3.   MDVP Fo(Hz) (Average vocal fundamental frequency): The fundamental frequency itself is a key factor in the prediction.
4.   MDVP Flo(Hz) (Minimum vocal fundamental frequency): The minimum frequency is another important measure.
5.   D2: A nonlinear dynamical complexity measure, indicating the presence of complex patterns in the voice of individuals with Parkinson's.

### **Other Notable Features:**

1.   spread2 and RPDE: These features capture additional aspects of non-linear variation and dynamical complexity.
2.   NHR (Noise-to-Harmonics Ratio) and HNR (Harmonics-to-Noise Ratio): These features reflect the noise components in the voice, which are important but less so than frequency-related features.

### **Insights:**
1.   PPE and other frequency-related features (MDVP:Fo(Hz), MDVP:Flo(Hz)) are the most critical for distinguishing between healthy individuals and those with Parkinson's disease.
2.   Non-linear dynamical features (like spread1, D2, and RPDE) are also significant, suggesting that the complexity of voice patterns is a key indicator of the disease.
3.   Noise-related features (like NHR and HNR) have some importance, indicating that increased noise in voice can be associated with Parkinson's.

```python
# Calculate the correlation matrix
corr_matrix_updrs = data_updrs.corr()

# Focus on the correlation of features with the target variables: motor_UPDRS and total_UPDRS
corr_with_motor_updrs = corr_matrix_updrs['motor_UPDRS'].sort_values(ascending=False)
corr_with_total_updrs = corr_matrix_updrs['total_UPDRS'].sort_values(ascending=False)

# Display the correlations
corr_with_motor_updrs, corr_with_total_updrs
```

### **Correlation Analysis Insights:**
* Strongest Correlations:

 * motor_UPDRS and total_UPDRS are highly correlated with each other (0.95), which is expected as they are related measures.

 * PPE (Pitch Period Entropy) and RPDE (Recurrence Period Density Entropy) show small but notable positive correlations with both motor_UPDRS and total_UPDRS.
 * Various shimmer and jitter measurements have smaller positive correlations with the UPDRS scores.

* Negative Correlations:
 * HNR (Harmonics-to-Noise Ratio) has a negative correlation with both UPDRS scores, suggesting that a lower HNR (indicating noisier voice) might be associated with higher UPDRS scores.

```python
from sklearn.ensemble import RandomForestRegressor

# Define the features and target variables
X = data_updrs.drop(columns=['motor_UPDRS', 'total_UPDRS'])
y_motor = data_updrs['motor_UPDRS']
y_total = data_updrs['total_UPDRS']

# Encode the categorical variable 'sex' if it's not numeric
if X['sex'].dtype == 'object':
    X['sex'] = X['sex'].astype('category').cat.codes

# Split the data into training and test sets
X_train_motor, X_test_motor, y_train_motor, y_test_motor = train_test_split(X, y_motor, test_size=0.3, random_state=42)
X_train_total, X_test_total, y_train_total, y_test_total = train_test_split(X, y_total, test_size=0.3, random_state=42)

# Standardize the features (important for tree-based models if we later compare with linear models)
scaler = StandardScaler()
X_train_motor = scaler.fit_transform(X_train_motor)
X_test_motor = scaler.transform(X_test_motor)
X_train_total = scaler.fit_transform(X_train_total)
X_test_total = scaler.transform(X_test_total)

# Train a Random Forest model
rf_motor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_motor.fit(X_train_motor, y_train_motor)

rf_total = RandomForestRegressor(n_estimators=100, random_state=42)
rf_total.fit(X_train_total, y_train_total)

# Get feature importances
importance_motor = rf_motor.feature_importances_
importance_total = rf_total.feature_importances_

# Get feature names
feature_names = X.columns

# Combine the feature importances with the feature names
feature_importance_motor = pd.DataFrame({'Feature': feature_names, 'Importance': importance_motor})
feature_importance_total = pd.DataFrame({'Feature': feature_names, 'Importance': importance_total})

# Sort the feature importances
feature_importance_motor = feature_importance_motor.sort_values(by='Importance', ascending=False)
feature_importance_total = feature_importance_total.sort_values(by='Importance', ascending=False)

feature_importance_motor, feature_importance_total
```

**Motor UPDRS:**

* Top Features:
 * subject#: This is the most important feature, indicating that the specific subject (which might be capturing individual-specific characteristics) is a strong predictor of the motor UPDRS score.
 * age: Age is the second most important feature, which is consistent with the understanding that Parkinson's disease symptoms often worsen with age.
 * test_time: The time since the subject's recruitment into the study also has a notable impact, possibly reflecting the progression of the disease over time.
 * DFA (Detrended Fluctuation Analysis): This feature contributes moderately to the prediction of motor UPDRS.

**Total UPDRS:**

* Top Features:
 * subject#: Similar to motor UPDRS, this is the most important feature for predicting total UPDRS.
 * age: Age again plays a significant role, reflecting the relationship between age and the severity of Parkinson's symptoms.
 * test_time and DFA: These features continue to be important for predicting the total UPDRS score.
 * HNR (Harmonics-to-Noise Ratio): HNR has some importance in predicting total UPDRS, albeit less than other top features.

**Insights:**
* subject# and age are the most significant predictors for both motor and total UPDRS scores.
* Voice-related features (e.g., jitter, shimmer, NHR, HNR) have relatively low importance compared to demographic and time-related features, but they still contribute to the model.
